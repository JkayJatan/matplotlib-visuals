```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
dhoni = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years,kohli,label="Virat kohli")
plt.plot(years,dhoni,label="M.S Dhoni")
plt.title("Performance Comparison")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.show()
```

```python
#format string
#eg 'ro--' , r=red,o=circle,--=doted
plt.plot(years,kohli,'ro--',label="Virat kohli")
plt.plot(years,dhoni,'bo--',label="M.S Dhoni")
plt.title("Performance Comparison")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.show()
```
