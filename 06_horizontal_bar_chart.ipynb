```python
import matplotlib.pyplot as plt
```

```python
languages = ['Python', 'JavaScript', 'Java', 'C++', 'C#', 'PHP']
popularity = [95, 88, 80, 70, 65, 50]
```

```python
plt.barh(languages, popularity, color='yellow')
plt.title("Programming Language Popularity (Example)", fontsize=14)
plt.xlabel("Popularity Score")
plt.ylabel("Languages")
plt.show()
```
